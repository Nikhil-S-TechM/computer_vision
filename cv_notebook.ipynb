#### Do all the necessary imports

```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

#### Read image and display its dimensions

```python
img1 = cv2.imread('./data/bicycle1.jpg')
img2 = cv2.imread('./data/bicycle2.jpg')

print('image 1 dimensions = ', img1.shape)
print('image 2 dimensions = ', img2.shape)
```

#### Convert to grayscale and resize the image

```python
img1_gray = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
img2_gray = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)

img1_gray = img1_gray[50:410,:]
print(img1_gray.shape)
print(img2_gray.shape)

fig = plt.figure(figsize=(20, 20))
fig.add_subplot(1,2,1)
plt.imshow(img1_gray, cmap='gray')
fig.add_subplot(1,2,2)
plt.imshow(img2_gray, cmap='gray')
```

#### Blend the 2 images 

```python
def blend(image1, image2, blend_ratio):
    blended_image = blend_ratio * image1 + (1-blend_ratio) * image2
    return blended_image

blended_image = blend(img1_gray, img2_gray, 0.8)
plt.imshow(blended_image, cmap='gray')
```
